```python
import numpy as np
import compGP

%matplotlib inline
```

```python
X=np.random.uniform(-3.,3.,(20,1))
Y=np.sin(X)+np.random.randn(20,1)*0.05
testX=np.linspace(-3,3)[None].T

m=compGP.RBF([0,0, 0.8151,1.8037 ,.0025])
```

```python
prior=m.prior(testX)
prior.plot_samples(nsamples=10)
```

```python
prior=m.prior(testX)
prior.plot_samples(summed=False,nsamples=10)
```

```python
posterior=m.posterior(X,Y,testX)
posterior.plot_samples(train=(X,Y),nsamples=10)
```

```python
posterior.plot_samples(summed=False,train=(X,Y),nsamples=10)
```

```python
posterior.plot_mean(train=(X,Y))
```

```python
posterior.plot_mean(summed=False,train=(X,Y))
```
